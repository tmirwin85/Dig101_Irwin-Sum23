# The Land of Little Rain

## Introduction

California literature includes extensive writings on the topic of the environment, American settlers, and Native peoples. The common theme of aridity unites all three. Mary Hunter Austin was an early nature writer of the American Southwest, best known for her book The Land of Little Rain, published in 1903. She came to the San Joaquin Valley with her family in 1888 and subsequently wrote about her experiences. Her writing reflects her interests in the Mojave desert, the aridity of the California environment, debates around California water rights, and Native American rights. Austin’s work today reveals changing attitudes at the turn of the century around the environment, the place of Native peoples, and conservation. An examination of The Land of Little Rain may indicate larger trends in early 20th century California literature and the emerging conservation movement, particularly as it regards deserts.

## Research Questions

While people have written about Austin and California nature writing, it is not a topic typically approached through digital technologies. With this in mind I asked two things of my textual analysis:

1) Given the influence of Austin’s work, and her own evolution in thinking about the California environment, what can The Land of Little Rain tell us about changing attitudes towards the environment, and the relationship of people to nature, at the turn of the century?

2) What can textual analysis targeting the frequency of language signal about the text’s core themes; the supremacy and divinity of nature, the negative consequences of the disconnect between humans and nature, and the positive consequences of the harmony between humans and nature?

Rawson and Muñoz speculate on alternative approaches to resolve the reductiveness of data cleaning in the humanities. However, these results are likely altered by the presence of punctuation and stop words. To get a better sense of the text, I can remove both.

```python
# For the stop words removal I will import nltk to my notebook
import nltk
from nltk.corpus import stopwords
stop = stopwords.words("english")
print(stop)
```

```python
# I can tokenize the text, which will split the text into words
rain_tokens = (no_rain.split())
```

```python
no_stops = [word for word in rain_tokens if word not in stopwords.words('english')]
print(no_stops)
```

```python
# Another thing is to lemmatize the text so that it is standardized, for this I need to import spaCy
import spacy
nlp = spacy.load("en_core_web_sm", disable=["ner", "textcat"])
```

```python
# spaCy expects a string as input, so I will use .join to force our list into a string  
words = ' '.join(no_stops)
words
```

```python
doc = nlp(words)
doc
```

```python
# We can now get the lemma of any word using the .lemma_ attribute
# notice this is "lose" not "lost"
doc[16].lemma_
```

```python
# In Notebook 2-1 there is a function to do all of this
# I already turned the tokens back into a string and processed it into doc, I now need to return the doc into tokens
def lemmatize(tokens):
    return [token.lemma_ for token in tokens]
```

```python
lemmas = lemmatize(doc)
print(lemmas)
```

The above steps mostly prepare my text to run through more meaningful tools that can help me to count the most frequent words and visualize the data. To do this I will need to set the text up as a dataframe.

```python
# I want to set this up as a dataframe
rain_nonumfreq = Counter(lemmas)
rain_nonumfreq
```

```python
# What are the 100 most frequent words?
rain_nonumfreq.most_common(100)
```

When I look at the list, I see that quotation marks make it into the second slot for most frequent words. While I didn't mean to leave punctuation like quotation marks in the text, it is interesting to consider the implications here. Quotation marks in writing have various meanings; they can indicate sarcasm, uncertainty, and dialogue. When I recall Rawson and Muñoz on data cleaning, these are the kinds of nuances lost when sorting for stop words. If I filter out punctuation that can indicate tone, will I dimish my ability to analyze the text and infer meaning? Leaving the quotation marks in (accidentally) may have given me insight into structure and tone that I didn't anticipate.

```python
rain_df = pd.DataFrame(rain_nonumfreq.most_common(100),
                        columns = ["Word", "Frequency"])
rain_df
```

Now I have this information in a dataframe. Before I proceed to the next task, I see another artifact from the punctuation removal; the 's indicating possession or as a contraction. On one hand it's probably a useless artifact of my earlier processing. On the other hand, if it does indicate possession, as of the land, this fits a settler mindset. When I return to my research questions, I wonder if the frequency of the possessive means that nature writers were seeking ownership of the environment even as they claimed (double entendre here) to protect/preserve it. The realtionship of people to nature was one that ranged from stewardship to ownership. Kevin Bruyneel has argued that a part of settler memory is white settler masculinity, to which white settler women often contributed. It focused on protection of the domestic realm, which put settlers at odds with Indigenous peoples and the environment as perceived obstacles. In this case, using digital tools to target the frequency of language (punctuation) actually reveals a subtext to the stories' themes - man's supremacy over nature.
